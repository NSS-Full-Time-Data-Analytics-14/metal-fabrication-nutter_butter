```python
j = jobs[['jmp_job_id','jmp_customer_organization_id']]
```

```python
so = sales_orders[['omp_sales_order_id','omp_full_order_subtotal_base']]
```

```python
soj = sales_order_job_links[['omj_sales_order_id','omj_job_id']]
```

```python
ssoj_merge = pd.merge(so, soj, 
        left_on = 'omp_sales_order_id', right_on = 'omj_sales_order_id', 
                how = 'left')
```

```python
s_soj_j_merge = pd.merge(j, ssoj_merge, 
        left_on = 'jmp_job_id', right_on = 'omj_job_id', 
                how = 'left')
```

# *** Top Revenue By Company Barplot

```python
ax = sns.barplot(top_revenue_by_customer_organization, x="omp_customer_organization_id", y="omp_full_order_subtotal_base", estimator="sum", errorbar=None)
plt.title('2023 and 2024 Total Revenue')
plt.ticklabel_format(style='plain', axis='y')
plt.xlabel("Organization ID")
plt.ylabel("Total Revenue in Dollars")
ax.patch.set_alpha(0.0)  # Set the axes background transparency
plt.gcf().patch.set_alpha(0.0)  # Set the figure background transparency
plt.xticks(rotation='vertical')
plt.savefig('top_revenue_by_org_id.jpg')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x:,.0f}'))
```
